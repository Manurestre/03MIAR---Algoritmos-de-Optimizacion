*   Se precisa coordinar el doblaje de una película. Los actores del doblaje deben coincidir en las tomas en las que sus personajes aparecen juntos en las diferentes tomas. Los actores de doblaje cobran todos la misma cantidad por cada día que deben desplazarse hasta el estudio de grabación independientemente del número de tomas que se graben. No es posible grabar más de 6 tomas por día. El objetivo es planificar las sesiones por día de manera que el gasto por los
servicios de los actores de doblaje sea el menor posible. Los datos son:

* Número de actores: 10
* Número de tomas : 30









                                        

```
# 1. Para leer el archivo desde google colab se necesitan permisos
from google.colab import auth
auth.authenticate_user()

# 2. Credenciales de google sheets
import gspread
from google.auth import default
creds, _ = default()

# 3. Autorización de conexión
gc = gspread.authorize(creds)

#4. Conexión 
worksheet = gc.open('Datos problema doblaje(30 tomas, 10 actores)').sheet1

#5. Trae todos los registros de la hoja de Sheets
rows = worksheet.get_all_values()

# 6. Usando Pandas, se crea el data frame para manipular los datos.
import pandas as pd
df = pd.DataFrame.from_records(rows)
#Quito encabezados y espacios en blanco para solo quedarme con los datos de tomas por cada actor en un nuevo dataframe
new_df = df.iloc[2:32,1:11]
new_df
```

```
# Crea una lista de días y una lista vacía para cada día
dias = []
for i in range(new_df.shape[1] - 1):
    dias.append([])

# Itera a través de cada fila (cada toma)
for i in range(new_df.shape[0]):
    # Itera a través de cada columna (cada actor)
    for j in range(new_df.shape[1] - 1):
        # Si el actor participa en la toma, agrega al actor al día correspondiente
        if df.iloc[i, j] == 1:
            dias[j].append(new_df.iloc[i, -1])

# Encuentra el día con el menor número de actores
min_actores = float("inf")
dia_min = -1
for i, dia in enumerate(dias):
    if len(dia) < min_actores:
        min_actores = len(dia)
        dia_min = i

# Imprime el día con el menor número de actores
print("El día con el menor número de actores es el día", dia_min + 1)
```

(*)Diseña un algoritmo que mejore la complejidad del algortimo por fuerza bruta. Argumenta porque crees que mejora el algoritmo por fuerza bruta

Respuesta

```
'''Para implementar un algortimo que lleve a cabo esta solución, opté por la opción de programación
dinámica con un algortimo llamado Floy-Warshall, lo que permite resolver el problema por los caminos más cortos con
un grafo ponderado con tiempo de ejecución O(n^2), y no mejora su coste computacional vs fuerza bruta'''
def numero_de_tomas(df):
    n_tomas = df.shape[0]
    n_actores = df.shape[1]
    
    tomas_seleccionadas = [0] * n_tomas
    for j in range(n_actores):
        for i in range(n_tomas):
            if df.iloc[i, j] == '1':
                tomas_seleccionadas[i] = 1
                break
    
    return sum(tomas_seleccionadas)

print("El número mínimo de tomas necesarias es:", numero_de_tomas(new_df))

```
